### Графики по видам опухолей

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from matplotlib import cm
import numpy as np
```

#### 1. Количество образцов в зависимости от типа опухоли

```python
df_files = pd.read_csv('../data/stats/cancer_type_files_count.csv', sep=';')

colors = cm.inferno_r(np.linspace(.4, .8, 33))

ax = df_files.plot(
  kind='bar',
  x='Type',
  y='Count',
  legend=False,
  figsize=(15, 8),
  width=0.8,
  color=colors,
  title='Количество образцов в зависимости от типа опухоли',
  xlabel='Тип опухоли',
  ylabel='Количество файлов'
) 

for bar in ax.patches:
  b = bar.get_bbox()

  ax.annotate(
    bar.get_height(),
    ((b.x0 + b.x1) / 2, b.y1),
    ha='center',
    va='bottom',
    fontsize=10
  )
```

#### 2. Количество мутаций в зависимости от типа опухоли

```python
df_cancers = pd.read_csv('../data/stats/cancer_type_mutations_count.csv', sep=';')

colors = cm.inferno_r(np.linspace(.4, .8, 33))

ax = df_cancers.plot(
  kind='bar',
  x='Type',
  y='Count',
  legend=False,
  figsize=(20, 8),
  width=0.8,
  color=colors,
  title='Количество мутаций в зависимости от типа опухоли',
  xlabel='Тип опухоли',
  ylabel='Количество мутаций'
) 

for bar in ax.patches:
  b = bar.get_bbox()

  ax.annotate(
    bar.get_height(),
    ((b.x0 + b.x1) / 2, b.y1),
    ha='center',
    va='bottom',
    fontsize=10
  )
```

#### 3. Количество мутаций в зависимости от типа опухоли с распределением по видам
