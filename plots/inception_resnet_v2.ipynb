### Графики зависимости метрик от эпох для теренировочных и валидационных данных модели InceptionResNetV2

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
```

```python
df1 = pd.read_csv('../history/inception_resnet_v2/dataset_1/metrics_dataset_1_04.05.2025-12.11.csv', sep=';')
df2 = pd.read_csv('../history/inception_resnet_v2/dataset_2/metrics_dataset_2_04.05.2025-14.29.csv', sep=';')
```

```python
plt.plot(df1['f1_macro'])
plt.plot(df1['val_f1_macro'])

plt.title('График зависимости F-меры (F1-score) с макро-усреднением от эпохи', fontsize=10)
plt.legend(['Тренировочные данные', 'Валидационные данные'], loc='lower right')

plt.ylabel('F-мера (F1-score)')
plt.xlabel('Эпоха')
```

```python
plt.plot(df1['f1_micro'])
plt.plot(df1['val_f1_micro'])

plt.title('График зависимости F-меры (F1-score) с микро-усреднением от эпохи', fontsize=10)
plt.legend(['Тренировочные данные', 'Валидационные данные'], loc='lower right')

plt.ylabel('F-мера (F1-score)')
plt.xlabel('Эпоха')
```

```python
plt.plot(df1['precision'])
plt.plot(df1['val_precision'])

plt.title('График зависимости точности (precision) от эпохи', fontsize=10)
plt.legend(['Тренировочные данные', 'Валидационные данные'], loc='lower right')

plt.ylabel('Точность (precision)')
plt.xlabel('Эпоха')
```

```python
plt.plot(df1['recall'])
plt.plot(df1['val_recall'])

plt.title('График зависимости полноты (recall) от эпохи', fontsize=10)
plt.legend(['Тренировочные данные', 'Валидационные данные'], loc='lower right')

plt.ylabel('Полнота (recall)')
plt.xlabel('Эпоха')
```

```python
plt.plot(df1['loss'])
plt.plot(df1['val_loss'])

plt.title('График зависимости функции потерь от эпохи', fontsize=10)
plt.legend(['Тренировочные данные', 'Валидационные данные'], loc='upper right')

plt.ylabel('Функция потерь')
plt.xlabel('Эпоха')
```

```python
df_matrix_1 = pd.read_csv('../history/inception_resnet_v2/dataset_1/matrix_dataset_1_04.05.2025-14.23.csv')
df_matrix_1 = df_matrix_1.drop(['Unnamed: 0'], axis=1)

labels = ['ACC', 'BLCA', 'BRCA', 'CESC', 'CHOL', 'COAD', 'DLBC', 'ESCA', 'GBM', 'HNSC', 'KICH', 'KIRC', 'KIRP', 'LAML', 'LGG', 'LIHC', 'LUAD', 'LUSC', 'MESO', 'OV', 'PAAD', 'PCPG', 'PRAD', 'READ', 'SARC', 'SKCM', 'STAD', 'TGCT', 'THCA', 'THYM', 'UCEC', 'UCS', 'UVM']

matrix_1 = df_matrix_1.to_numpy()

plt.figure(figsize = (15, 15))

sns.set(font_scale=1.4)
sns.heatmap(matrix_1, cmap="Blues", annot=True, annot_kws={"size": 14}, xticklabels=labels, yticklabels=labels, fmt='d')

plt.ylabel('Правильные значения')
plt.xlabel('Предсказанные значения')
plt.show()
```
